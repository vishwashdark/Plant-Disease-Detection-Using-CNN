```python
import tkinter as tk
from tkinter import filedialog
from PIL import Image, ImageTk
import keras
import numpy as np

# Load the saved model
model = keras.models.load_model("D:/model_file.h5")

# Create object
root = tk.Tk()

# Adjust size
root.geometry("1000x1000")
root.configure(bg="green")

# Add image file
bg = ImageTk.PhotoImage(file="C:/Users/sharm/OneDrive/Desktop/plant1.png")

# Create Canvas
canvas1 = tk.Canvas(root, width=root.winfo_screenwidth(), height=root.winfo_screenheight(), bg='light green')
canvas1.pack(fill="both", expand=True)

# Display image
canvas1.create_image(220, 50, image=bg, anchor="nw")

# Add Text
canvas1.create_text(850, 30, text="PLANT DISEASE DETECTION", font=('Times', 24))

def open_page2():
    root.destroy()
    my_w = tk.Tk()
    my_w.geometry("1200x600")
    my_w.title('KASHMIRI APPLE LEAF DISEASE DETECTION')
    my_font1 = ("Helvacita", 18, 'bold')
    l1 = tk.Label(my_w, text='Add leaf Image', height=10, width=80, font=my_font1)
    l1.grid(row=1, column=1)
    
    def upload_file():
        global img
        f_types = [('Jpg Files', '*.jpg'),("JPEG Files" ,"*.jpeg.jpg")]
        filename = filedialog.askopenfilename(filetypes=f_types)
        img = ImageTk.PhotoImage(file=filename)
        img_pil = Image.open(filename).resize((64, 64))
        img_arr = np.array(img_pil) / 255.0
        input_arr = np.expand_dims(img_arr, axis=0)
        predictions = model.predict(input_arr)
        
        if predictions[0] > 0.005:
            print("Healthy Leaf")
            a = tk.Label(my_w, text="Leaf Report")
            a.grid(columnspan=3, column=0, row=3)
            inst = tk.Label(my_w, text="   HEALTHY LEAF   ")
            inst.grid(columnspan=3, column=0, row=4)
        else:
            print("Normal")
            a = tk.Label(my_w, text="Leaf Report")
            a.grid(columnspan=3, column=0, row=3)
            inst = tk.Label(my_w, text="    ROT LEAF    ")
            inst.grid(columnspan=3, column=0, row=4)
    
    b1 = tk.Button(my_w, text='Upload File', width=20, command=upload_file)
    b1.grid(row=2, column=1)
    my_w.mainloop()

# Create Buttons
button1 = tk.Button(root, text="Apple", bg="RED", font=('Arial Bold', 16), height=2, width=10, command=open_page2)
button2 = tk.Button(root, text="Rice", bg="PINK", font=('Arial Bold', 16), height=2, width=10)

# Display Buttons
button1_canvas = canvas1.create_window(400, 730, anchor="c", window=button1)
button2_canvas = canvas1.create_window(800, 730, anchor="c", window=button2)

# Execute tkinter
root.mainloop()
```
